trying RF with different number of trees

```
from pyspark.ml.classification import LogisticRegression, RandomForestClassifier
from pyspark.ml.feature import StandardScaler
```

```
training_df,validation_df, test_df = numeric_df.randomSplit([0.6,0.3,0.1])
```

```
training_df.printSchema()
```

```
va = VectorAssembler(). \
      setInputCols(training_df.columns[1:]). \
      setOutputCol('features')
```

```
#default
rf = RandomForestClassifier(maxBins=10000)
rf15 = RandomForestClassifier(numTrees=15, maxBins=10000)
rf25 = RandomForestClassifier(numTrees=25, maxBins=10000)
```

```
rf_pipeline = Pipeline(stages=[va, rf]).fit(training_df)
rf_pipeline_15 = Pipeline(stages=[va, rf15]).fit(training_df)
rf_pipeline_25 = Pipeline(stages=[va, rf25]).fit(training_df)
```

```
rf_pipeline.transform(validation_df)
rf_pipeline_15.transform(validation_df)
rf_pipeline_25.transform(validation_df)
```

```
from pyspark.ml.evaluation import BinaryClassificationEvaluator
bce = BinaryClassificationEvaluator()
```

```
print(bce.evaluate(rf_pipeline.transform(validation_df)))
print(bce.evaluate(rf_pipeline_15.transform(validation_df)))
print(bce.evaluate(rf_pipeline_25.transform(validation_df)))
```
